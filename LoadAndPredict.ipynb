```python
# LOAD MODEL
n = "5"
print("Loading model #"+n)
model = load_model("./Models/neuralmon_model_"+n+".hdf5")
```

```python
# LOAD IMAGES
folder = "RealWorldPokemon"
images = []
for filename in os.listdir(folder):
    if filename[0] != '.':
        img = Image.open(folder + "/" + filename)
        images.append(np.array(img))
ims = np.array(images)
print(ims.shape)
```

```python
ims_normalized = ims.astype('float32') / 255
ims_normalized -= np.mean(ims_normalized)
```

```python
scores = model.predict(ims_normalized)
```
